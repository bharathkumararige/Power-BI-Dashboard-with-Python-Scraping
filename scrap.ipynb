```python
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
```

```python
##def scrape_book():
url = "http://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"}
res = requests.get(url, headers = headers)
```

```python
 res = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(res.content, "html.parser")
books = soup.find_all("article", class_="product_pod")
```

```python
data = []
for book in books:
    title = book.find("h3").find("a")["title"]
    links = book.find("h3").find("a")["href"]
    price = book.find("p", class_="price_color").text.strip().replace("£", "")
    category = "General"
    date = datetime.now().date()
    data.append([title, links ,float(price), category, date])
```

```python
print(links)
```

```python
df = pd.DataFrame(data, columns=["Title","Link_of_book", "Price", "Category", "Date"])
df.dropna(inplace=True)
```

```python
print(df)
```

```python
df.to_csv("booksc.csv", index=False)
print(f"Data saved as CSV at {datetime.now()}")
# Verify
print(f"File exists: {os.path.exists('booksc.csv')}")
```
